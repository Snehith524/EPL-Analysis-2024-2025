```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("epl_player_stats_24_25.csv")

# Filter disciplined players: >1710 minutes, 0 yellow cards, 0 red cards
disciplined_players = df[
    (df["Minutes"] > 1710) &
    (df["Yellow Cards"] == 0) &
    (df["Red Cards"] == 0)
].sort_values(by="Minutes", ascending=False)

# Plot
plt.figure(figsize=(12, 7))
sns.barplot(x="Minutes", y="Player Name", data=disciplined_players, palette="pastel")
plt.title("Most Disciplined Players (0 Cards & >1710 Minutes) - EPL 2024/25")
plt.xlabel("Minutes Played")
plt.ylabel("Player Name")
plt.tight_layout()
plt.show()

# Display table
disciplined_players[["Player Name", "Club", "Position", "Minutes", "Yellow Cards", "Red Cards"]]
```
